```python
#Cody Rorick
#purpose: To find the return and losses associated with lower and higher risk leveraged bull call spreads
#inputs: Yahoo finance stock market data
#outputs: Bar plot of bull spread risk/reward profile
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
def getData(symbol):

    tk = yf.Ticker(symbol)
    # Expiration dates
    exps = tk.options
    print('A list of all contract expiration dates: ')
    print(*exps)
    expDate = input('\nWhich date would you like to analyze in yyyy-mm-dd format: ' )
    
    #creating df for options
    options = pd.DataFrame()
    
    #getting options data for specific date
    options = tk.option_chain(expDate).calls
    options['expirationDate'] = expDate
        
    #getting the average strike price
    options[['bid', 'ask', 'strike']] = options[['bid', 'ask', 'strike']].apply(pd.to_numeric)
    options['mark'] = (options['bid'] + options['ask']) / 2 
    
    # Drop unnecessary columns
    options = options.drop(columns = ['inTheMoney', 'contractSymbol', 'volume', 'openInterest', 'impliedVolatility', 'contractSize', 'currency', 'change', 'percentChange', 'lastTradeDate', 'lastPrice'])
    
    #adding ticker and current stock price to dataframe
    options['symbol'] = symbol
    options['current price'] = tk.info['currentPrice']
    
    #price underlying asset has to reach to break even
    options['breakeven'] = options['mark'] + options['strike']
    

    return options
```

```python
def callFinder(df, cushionPercent):
    #Target price will determines where you want to buy the call and the first leg of the spread to start
    targetPrice = input(f"current price is: {df.loc[0, 'current price']}, what is your price target for: {df.loc[0, 'expirationDate']}: ")
    
    #The true projected price of the stock (where you break even) is the premium plus the strike price
    truePrices = list(df.loc[:, 'breakeven'])
    
    #The target breakeven is the point where your target price equals the prem+strike
    targetBreakeven = min(truePrices, key=lambda x:abs(x-float(targetPrice)))
    
    #How much cushion do you want to put in your forecast
    riskAdjustedPrice = min(truePrices, key=lambda x:abs(x-(targetBreakeven*(1-cushionPercent))))
    
    #only grab the data at or above the cushion
    df = df.loc[df['breakeven'] >= riskAdjustedPrice, :]
    df = df.reset_index(drop = True)

    #Beyond the initial strike that is bought, all calls to be sold should be OTM from target price
    firstLeg = df.iloc[[0]]
    secondLeg = df.loc[df['strike'] > float(targetPrice), :]
    df = pd.concat([secondLeg, firstLeg], ignore_index = True)
    df = df.sort_values(by=['breakeven'])

    return df
```

```python
def spreadFinder(df):
    bullSpread = pd.DataFrame(columns = ['current price', 'symbol', 'buyStrike', 'buyStrikePrem', 'sellStrike', 'sellStrikePrem', 'maxProf', 'maxLoss', 'breakevenPrice', 'date'])
    for i in range(1, int(len(df)/2)):
        bullSpread.loc[i-1, 'buyStrike'] = df.loc[0, 'strike']
        bullSpread.loc[i-1, 'buyStrikePrem'] = df.loc[0, 'mark']
        bullSpread.loc[i-1, 'sellStrike'] = df.loc[i, 'strike']
        bullSpread.loc[i-1, 'sellStrikePrem'] = df.loc[i, 'mark']
        bullSpread.loc[i-1, 'maxLoss'] = (df.loc[0, 'mark'] - df.loc[i, 'mark']) * 100
        bullSpread.loc[i-1, 'maxProf'] = (df.loc[i, 'strike'] - df.loc[0, 'strike'] - (df.loc[0, 'mark'] - df.loc[i, 'mark'])) * 100
        bullSpread.loc[i-1, 'breakevenPrice'] = df.loc[0, 'strike'] + (df.loc[0, 'mark'] - df.loc[i, 'mark'])
        bullSpread.loc[i-1, 'date'] = df.loc[0, 'expirationDate']
        bullSpread.loc[i-1, 'symbol'] = df.loc[0, 'symbol']
        bullSpread.loc[i-1, 'current price'] = df.loc[0, 'current price']
    return bullSpread
```

```python
def strikeBarPlotter(df):
    #sort the options data and set up the information for the axes
    df['breakevenPrice'] = df['breakevenPrice'].apply(lambda x: round(x, 2))
    df = df.sort_values(by='breakevenPrice')
    X = list(df['breakevenPrice'])
    X_axis = np.arange(len(X))
    width = 0.15
    vals = []
    bars = []
    
    #the % return for each %OTM call will be the height of the bar
    vals.append(list(df['maxProf'].values))
    vals.append(list(df['maxLoss'].values))
        
    #the bars will be different shades on greyscale
    for i in range(0, len(vals)):
        bars.append(plt.bar(X_axis + i * width, vals[i], width, color = str((1/(i+1.125)))))
    
    #set up the chart axes labeling
    plt.xlabel('Break Even Price ($)') 
    plt.ylabel('Risk/Reward Profile ($)') 
    plt.title(f"Bull call spreads for: {df.loc[0, 'symbol']}, Date: {df.loc[0, 'date']}, Current Price: ${df.loc[0, 'current price']}")
    
    #generate the plot with ticks at every spread
    plt.xticks(X_axis+width, X, fontsize = 8) 
    plt.legend(bars, ['Max Profit', 'Max Loss']) 
    plt.show() 
```

```python
#choosing the stock and percent under the breakeven you want to be
stock = input('Which stock do you want to see bull spreads?: ')
riskAdjustment = float(input('Risk adjustment factor, how aggressive do you want to be with a price target (-0.1 aggressive breakeven to 0.1 conservative breakeven): '))
callData = callFinder(getData(stock), riskAdjustment)
callData = callData.reset_index(drop = True)
bullSpread = spreadFinder(callData)
```

```python
strikeBarPlotter(bullSpread)
```
